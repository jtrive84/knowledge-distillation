Knowing the number of parameters in your models is crucial for two reasons: The parameter count directly impacts computational resources, determining memory requirements, training time, and deployment costs. Perhaps more importantly, models with excessive parameters relative to dataset size risk overfitting, essentially memorizing training data rather than learning generalizable patterns. This balance between parameter count and training data is fundamental to achieving good generalization capability.

Models can have two types of parameters:

- **trainable parameters**: 
    - Have `requires_grad=True` (this is the default for most parameters)
    - Are updated during backpropagation
    - Contribute to the model's learning capacity
    - Are included in optimizer updates
    - Consume memory for both parameter storage and gradient storage

    
- **Non-trainable Parameters**:
    - Have `requires_grad=False`
    - Are not updated during backpropagation
    - Still contribute to the forward pass computations
    - Are not included in optimizer updates
    - Only consume memory for parameter storage (no gradients)

    
For the very simple networks used in this module, all parameters will be trainable. But as we move to more advanced architectures, models will have a blend of trainable and non-trainable parameters.

Determining the number of model parameters is straightforward in PyTorch. Referring again to the `SimpleNetwork` architecture defined earlier:

```python

import torch
import torch.nn as nn

class SimpleNetwork(nn.Module):
    
    def __init__(self, p=0.0):
        
        super().__init__()

        self.model = nn.Sequential(
            nn.Flatten(),
            nn.Linear(in_features=28 * 28, out_features=512),
            nn.ReLU(),
            nn.Dropout(p=p),
            nn.Linear(in_features=512, out_features=10)
        )

    def forward(self, X):
        output = self.model(X)
        return output
```


<br>

To count the total number of parameters and the number of trainable parameters, create an instance of the model of interest and run the following:

```python

# Create model instance.
model = SimpleNetwork()

# Count total parameters.
total_params = sum(p.numel() for p in model.parameters())

# Count trainable parameters only.
trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)

print(f"Total parameters    : {total_params:,}")
print(f"Trainable parameters: {trainable_params:,}")
```


<br>

It is also important to be able to calculate the number of model parameters by analyzing the architecture. `SimpleNetwork` consists of two layers:

- The first layer accepts an input of size **784**, and steps it down to a vector of length **512**. 
- The second layer accepts an input of size **512**, and steps it down to a vector of length **10**. 

We then have:

- The first set of weights are contained within a 784x512 matrix. 
- The bias for the first layer has the same size as the output dimension of the first layer (512).
- The second set of weights are contained within a 512x10 matrix. 
- The bias for the second layer has the same size as the output dimension of the second layer (10).

Therefore, the total number of parameters is:


```python

784 * 512 + 512 * 10 + 512 + 10
```

<br>

This matches the result returned by PyTorch.
