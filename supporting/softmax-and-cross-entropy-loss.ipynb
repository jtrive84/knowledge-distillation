Softmax is a function that converts a vector of raw scores or logits into probabilities ("logits" refer to the raw, unnormalized output values produced by the last layer of a neural network before applying an activation function).
It is commonly used in classification tasks, particularly in the output layer of a neural network. It converts the raw logits into a probability distribution.

For a vector $z$ of length $n$, softmax is defined as:

$$
\sigma(z_i) = \frac{e^{z_i}}{\sum_{j=1}^{n} e^{z_j}},
$$

where:
- $e^{z_i}$ exponentiates each element.
- The denominator ensures all probabilities sum to 1.

<br>

If any of the $z_i$ is large, $e^{z_i}$ can become extremely large, potentially causing overflow errors in computation. For example:

```python
import numpy as np

z = np.array([1000, 2000, 3000])         # Large values
softmax = np.exp(z) / np.sum(np.exp(z))  # OverflowError
```

Since $e^{3000}$ is astronomically large, Python will struggle to handle it. The solution is to subtract the max value
of each sample instead of using $z$ directly:

$$
\sigma(z_i) = \frac{e^{z_i - \max(z)}}{\sum_{j=1}^{n} e^{z_j - \max(z)}}
$$

<br>

This shifts all values down without affecting the final probabilities (because shifting inside the exponent maintains relative differences).

### **Cross-Entropy Loss**

Cross-entropy loss is commonly used as the loss function in classification problems. It measures the difference between two probability distributions: 

- The **true labels** (can be one-hot encoded).  
- The **predicted probabilities** (from softmax).  

A lower cross-entropy loss means the predicted probabilities are closer to the true labels.

A key feature of cross entropy loss is that it rewards/penalizes the probabilities of correct classes only. The value is independent of how the remaining probability is split between the incorrect classes. 

For a single sample with \( C \) classes, the cross-entropy loss is give by

$$
L = - \frac{1}{n}\sum_{i=1}^{C} y_i \times \log(\hat{y_i}),
$$

where:
- $n$ is the batch size.
- $y_i$ is the true label (1 for the correct class, 0 otherwise).
- $\hat{y_i}$ is the predicted probability for class $i$ (from softmax).
- The $\log$ ensures the loss is large when the predicted probability is low for the correct class.


If we had a single vector of actual labels representing the index of the correct class (i.e., `yact` from above), simply compute the negative log of the probability at this index to get the cross entropy loss for that sample (since cross-entropy doesn't consider incorrect classes).


In PyTorch, if `CrossEntropyLoss` is used for the loss function, there is no need to include softmax after the final layer activations as this is done automatically. This is not the case when using `NLLLoss`. If using `NLLLoss`, softmax must be included.
